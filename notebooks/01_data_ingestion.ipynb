```python
import yfinance as yf 
import pandas as pd 

tickers = [
    "AAPL",
    "GOOGL",
    "MSFT",
    "VNM.VN",
    "HPG.VN"
]

all_dfs = []

for ticker in tickers:
    df = yf.download(
        ticker, 
        start="2021-01-01", 
        end="2024-01-05",
    )

    df.columns = [
            col[0] if isinstance(col, tuple) else col
            for col in df.columns
        ]

    df = df.reset_index()

    # thêm ticker
    df["ticker"] = ticker

    all_dfs.append(df)
```

```python
final_pd = pd.concat(all_dfs)
spark_df = spark.createDataFrame(final_pd)
```
